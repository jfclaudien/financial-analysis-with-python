#### Moyenne Mobile ou Glissante (Rolling et Expanding)

```python
walmart = pd.read_csv('./data/walmart_stock.csv', 
                      index_col= 'Date', 
                      parse_dates= True)
```

```python
walmart.head()
```

```python
walmart.Open.plot(figsize=(15, 5));
```

```python
walmart.rolling(window= 7).mean().head(15)
```

```python
walmart.Open.plot(figsize= (16, 6))
walmart.rolling(90).mean()['Close'].plot();
```

```python
walmart['Close 30 days MA'] = walmart.Close.rolling(window= 30).mean()
```

```python
walmart[['Close 30 days MA', 'Close']].plot(figsize= (16, 6));
```

```python
walmart.Close.expanding(min_periods= 1).mean().plot(figsize= (16, 6));
```

#### Indicateur Technique : "Bandes de Bollinger"
    Close 20 MA -> Moving Average
    Upper = 20 MA + 2 * std(20)
    Lower = 20 MA - 2 * std(20)
    Close

```python
walmart['Close 20 days MA'] = walmart.Close.rolling(window= 20).mean()
walmart['Upper'] = walmart['Close 20 days MA'] + 2 * walmart.Close.rolling(20).std()
walmart['Lower'] = walmart['Close 20 days MA'] - 2 * walmart.Close.rolling(20).std()
walmart[['Close', 'Close 20 days MA', 'Upper', 'Lower']].plot(figsize= (16, 6));
```

```python
walmart['Close 20 days MA'] = walmart.Close.rolling(window= 20).mean()
walmart['Upper'] = walmart['Close 20 days MA'] + 2 * walmart.Close.rolling(20).std()
walmart['Lower'] = walmart['Close 20 days MA'] - 2 * walmart.Close.rolling(20).std()
walmart[['Close', 'Close 20 days MA', 'Upper', 'Lower']].tail(200).plot(figsize= (16, 6));
```
